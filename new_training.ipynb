```python
total_score = deliveries.groupby(["match_id", "inning"]).sum()["total_runs"].reset_index()
total_score = total_score[total_score["inning"] == 1]

'''
1. Here first we performed groupby -> it groups the two columns as one like in sql
    Dataset has values like runs, overs etc. Here it combines all when grouped by
2. Then performing .sum() combines their internal values
    Like runs -> total_runs, overs -> total_overs
3. It finds the total_runs columns from the group. by performing [total_sum]
4. .reset_index() resets the index again from 1 to len(deliveries)'''
```

```python
total_score.head()
```

```python
match = matches.merge(total_score[["match_id","total_runs"]], left_on = "id", right_on = "match_id")
'''
Here panda merges `matches` and `total_score`
Here the merging criteria with left_on and right_on is:
matches.id == total_score.match_id'''
```

```python
teams = [
    "Chennai Super Kings",
    "Delhi Capitals",
    "Gujarat Titans",
    "Kolkata Knight Riders",
    "Lucknow Super Giants",
    "Mumbai Indians",
    "Punjab Kings",
    "Rajasthan Royals",
    "Royal Challengers Bengaluru",
    "Sunrisers Hyderabad"
]
```

```python
match["team1"] = match["team1"].str.replace("Royal Challengers Bangalore","Royal Challengers Bengaluru")
match["team2"] = match["team2"].str.replace("Royal Challengers Bangalore","Royal Challengers Bengaluru")

match["team1"] = match["team1"].str.replace("Delhi Daredevils","Delhi Capitals")
match["team2"] = match["team2"].str.replace("Delhi Daredevils","Delhi Capitals")

match["team1"] = match["team1"].str.replace("Kings XI Punjab","Punjab Kings")
match["team2"] = match["team2"].str.replace("Kings XI Punjab","Punjab Kings")

match["team1"] = match["team1"].str.replace("Deccan Chargers","Sunrisers Hyderabad")
match["team2"] = match["team2"].str.replace("Deccan Chargers","Sunrisers Hyderabad")
```

```python
match["team1"].unique()
```

```python
match = match[match["team1"].isin(teams)]
match = match[match["team2"].isin(teams)]
```

```python
match = match[["match_id","city","winner","total_runs"]]
match.head()
```

```python
deliveries_df = match.merge(deliveries, on = "match_id")
deliveries_df.head()
```

```python
deliveries_df = deliveries_df[deliveries_df["inning"] == 2]
deliveries_df.head()
```

```python
deliveries_df.columns
```

```python
deliveries_df["batting_team"] = deliveries_df["batting_team"].str.replace("Royal Challengers Bangalore","Royal Challengers Bengaluru")
deliveries_df["bowling_team"] = deliveries_df["bowling_team"].str.replace("Royal Challengers Bangalore","Royal Challengers Bengaluru")

deliveries_df["batting_team"] = deliveries_df["batting_team"].str.replace("Delhi Daredevils","Delhi Capitals")
deliveries_df["bowling_team"] = deliveries_df["bowling_team"].str.replace("Delhi Daredevils","Delhi Capitals")

deliveries_df["batting_team"] = deliveries_df["batting_team"].str.replace("Kings XI Punjab","Punjab Kings")
deliveries_df["bowling_team"] = deliveries_df["bowling_team"].str.replace("Kings XI Punjab","Punjab Kings")

deliveries_df["batting_team"] = deliveries_df["batting_team"].str.replace("Deccan Chargers","Sunrisers Hyderabad")
deliveries_df["bowling_team"] = deliveries_df["bowling_team"].str.replace("Deccan Chargers","Sunrisers Hyderabad")
```

```python
deliveries_df["batting_team"].unique()
```

```python
deliveries_df["current_score"] = deliveries_df.groupby("match_id")["total_runs_y"].cumsum()
deliveries_df.head()
```

```python
## find the runs left
deliveries_df["runs_left"] = deliveries_df["total_runs_x"] - deliveries_df["current_score"]
deliveries_df.head()
```

```python
deliveries_df["balls_left"] = 120 - (deliveries_df["over"]*6 + deliveries_df["ball"])
deliveries_df.head()
```

```python
deliveries_df.isnull().sum()
```

```python
deliveries_df["player_dismissed"] = deliveries_df["player_dismissed"].fillna(0)
deliveries_df["player_dismissed"] = deliveries_df["player_dismissed"].apply(lambda x:x if x == 0 else 1)
```

```python
## finding the wickets left
wickets = deliveries_df.groupby("match_id")["player_dismissed"].cumsum()
deliveries_df["wickets_left"] = 10 - wickets
deliveries_df.head()
```

```python
# finding how many wickets gone
deliveries_df["wickets"] = wickets
deliveries_df.head()
```

```python
deliveries_df.tail()
```

```python
# finding current Run Rate
deliveries_df["crr"] = (deliveries_df["current_score"]*6) / (120 - deliveries_df["balls_left"])
deliveries_df.head()
```

```python
## Finding Required Run Rate
deliveries_df["rrr"] = (deliveries_df["runs_left"]*6) / (deliveries_df["balls_left"] + 0.0001)
deliveries_df.tail()
```

```python
deliveries_df["r/w_ratio"] = deliveries_df["current_score"]/(deliveries_df["wickets"] + 0.00001)
deliveries_df.tail()
```

```python
def result(row):
    return 1 if row["batting_team"] == row["winner"] else 0
```

```python
deliveries_df["result"] = deliveries_df.apply(result, axis = 1)
deliveries_df.head()
```

```python
deliveries_df.columns
```

```python
print("city, total_runs_x, batting_team, bowling_team, runs_left, balls_left, wickets_left, crr, rrr, r/w_ratio, result")
```

```python
final_df = deliveries_df[["batting_team","bowling_team","city","runs_left","balls_left","wickets_left","total_runs_x","crr","rrr","r/w_ratio","result"]]
```

```python
final_df.head()
```

```python
final_df = final_df.sample(final_df.shape[0])
final_df.tail()
```

```python
final_df.isnull().sum()
```

```python
final_df.dropna(inplace = True)
```

```python
# splitting to x and y
# then train, test, split
x = final_df.drop("result",axis = 1)
y = final_df["result"]

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer(
    [("trf",OneHotEncoder(drop = "first"), ["batting_team","bowling_team","city"])
], remainder = "passthrough"
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps = [
    ("step1",trf),
    ("step2",RandomForestClassifier())
])
```

```python
pipe.fit(xtrain, ytrain)
```

```python
## predicting 
ypred = pipe.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ypred, ytest)
```
